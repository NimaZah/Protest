```python
# loading the data into a dataframe
df = pd.read_csv('https://raw.githubusercontent.com/joankovacs/mass_mobilization/main/cleaned_protests.csv')
#check the first rows of the dataframe
df.head()
df.drop(['sources', 'notes', 'protesteridentity'], axis=1, inplace=True)
df.head()
```

```python
#print the unique values of duration and assign them to unique.
unique = df['protest_length'].unique()
# sort the values of unique.
unique.sort()
unique
```

```python
# create a function to assign a category to the values of protest_length column. The function will check if the value 
# of the protest_length is less than or equal to 30, and if it is, then it returns '1 months'. If the value of the
# protest_length column is greater than 30 but less than or equal to 90, then it returns '1-3 months'. If 
# value of the protest_length column is greater than 90 but less than or equal to 180, then it returns
# '1-6 months'. If the value of the protest_length column is greater than 180, then it returns '>6 months'

def protest_length_category(protest_length):
    if protest_length <= 30:
        return '1 month'
    elif protest_length > 30 and protest_length <= 90:
        return '1-3 months'
    elif protest_length > 90 and protest_length <= 180:
        return '3-6 months'
    elif protest_length > 180:
        return '>6 months'
```

```python
# Apply the function to the protest_length column.
# df['protest_length_category'] = df.apply(protest_length_category, axis=1)
# df['protest_length_category']
# df.head()
df['protest_length_category'] = df['protest_length'].apply(protest_length_category)
df.head()
```

```python
# define the app
app = Flask(__name__)

# define the entry point
@app.route('/')
def index():
    return render_template('app.html')

# define the output page
@app.route('/submit', methods=['POST'])
def submit():
    ''' render home.html - page that is served at localhost that allows user to enter model scoring parameters'''
    title_text = 'Enter parameters for model scoring:'
    origin_text = 'Origin'
    destination_text = 'Destination'
    return render_template('index.html', title_text=title_text, origin_text=origin_text, destination_text=destination_text)

# define the output page
@app.route('/score', methods=['POST'])
def score():
    '''render score.html page that is served at localhost that displays the results of the scoring'''
    # get the inputs from the form
    origin = request.form.get('origin')
    destination = request.form.get('destination')
    
    # get the current working directory
    cwd = os.getcwd()
    
    # get the model
    model_file = cwd+'/trained_model_protests.pkl'
    
    # create a dataframe for the model scoring
    input_data = pd.DataFrame({'country': [origin], 'region': [destination]})
    
    # create a TabularPandas object for the input dataframe
    proc_input = TabularPandas(input_data, procs=[FillMissing, Categorify, Normalize],
                               cat_names=['country', 'region'], cont_names=[], y_names=[])
    
    # apply the TabularPandas object to the input dataframe to create the input dataloader
    test_dataloader = proc_input.test_dl(proc_input.X_test)
    
    # load the model
    learn2 = load_learner(path=model_file, test=test_dataloader)
    
    # predict the output
    pred = learn2.get_preds(ds_type=DatasetType.Test)[0][:, 1]
    pred = pred.numpy().flatten()
    pred = np.round(pred, 3)
    prediction = str(pred[0])
    
    # render the score.html page
    return render_template('score.html', prediction=prediction)

if __name__ == "__main__":
    app.run()
```

```python
# define the app
app = Flask(__name__)

# define the entry point
@app.route('/')
def index():
    return render_template('app.html')

# define the output page
@app.route('/submit', methods=['POST'])
def submit():
    ''' render home.html - page that is served at localhost that allows user to enter model scoring parameters'''
    country = request.form.get('country')
    region = request.form.get('region')
    protest_length = request.form.get('protest_length')
    year = request.form.get('year')
    month = request.form.get('month')
    day = request.form.get('day')
    week = request.form.get('week')
    protest_type = request.form.get('protest_type')
    return render_template('home.html',country=country,region=region,protest_length=protest_length,year=year,month=month,day=day,week=week,protest_type=protest_type)

# define the output page
@app.route('/output')
def output():
    ''' render output.html - page that is served at localhost that shows prediction results'''
    # pull input values from home.html
    country = request.args.get('country')
    region = request.args.get('region')
    protest_length = request.args.get('protest_length')
    year = request.args.get('year')
    month = request.args.get('month')
    day = request.args.get('day')
    week = request.args.get('week')
    protest_type = request.args.get('protest_type')
    # load the model
    MODEL_PATH = 'trained_model_protests.pkl'
    # load model using pickle
    model = pickle.load(open(MODEL_PATH, 'rb'))
    # create dataframe with user inputs
    df = pd.DataFrame(columns=['country','region','protest_length','protest_type','year','month','day','week'])
    df.loc[0] = [country,region,protest_length,protest_type,year,month,day,week]
    # make prediction
    y_pred_prob = model.predict_proba(df)
    y_pred_prob_value = y_pred_prob[0][1]
    # output page
    return render_template('output.html',prediction=y_pred_prob_value)
1
```
